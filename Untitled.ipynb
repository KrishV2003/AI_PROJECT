```python
import requests

# Define the coordinates for Pune, Maharashtra, India
lat = 18.5204
lon = 73.8567

# Build the API URL
url = f"https://api.co2signal.com/v1/latest?lat={lat}&lon={lon}"

# Make the API request (replace 'YOUR_API_KEY' with your actual API key)
headers = {
    'Authorization': 'Bearer 3vzeyoQxnf638'  # Your actual API key here
}

response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    print(data)  # Print the entire response data for debugging

    # Access the carbon intensity based on the output structure
    carbon_intensity = data['data']['carbonIntensity']  # Adjust this line if necessary
    print(f"Carbon Intensity for Pune: {carbon_intensity} gCO2/kWh")
else:
    print(f"Error: {response.status_code}, {response.text}")
```

```python
import pandas as pd
# Create a DataFrame to store the data
energy_consumption = 500  # Example energy consumption for CtrlS in kWh

# Create a DataFrame
data = pd.DataFrame({
    'carbon_intensity': [carbon_intensity],
    'energy_consumption': [energy_consumption]
})

# Display the DataFrame
print(data)
```

```python
from sklearn.linear_model import LinearRegression

# Since you only have one data point, just use that data point to fit the model
X = data[['carbon_intensity', 'energy_consumption']]
y = data['carbon_intensity']  # Carbon intensity as the target

# Initialize the model
model = LinearRegression()

# Fit the model with your single data point (for demonstration purposes)
model.fit(X, y)

# Since we only have one sample, use the same data for prediction
y_pred = model.predict(X)

# Display predictions
print(f"Predicted Carbon Intensity for CtrlS Data Center: {y_pred}")
```

```python
# Example: Predicting carbon footprint for CtrlS based on new data
new_data = pd.DataFrame({
    'carbon_intensity': [300],  # Example input for carbon intensity
    'energy_consumption': [500]  # Example input for energy consumption
})

predicted_intensity = model.predict(new_data)
print(f'Predicted Carbon Intensity for CtrlS Data Center: {predicted_intensity[0]} gCO2/kWh')

# Calculate the carbon footprint
carbon_footprint = new_data['energy_consumption'][0] * predicted_intensity[0]
print(f'Estimated Carbon Footprint for CtrlS Data Center: {carbon_footprint} grams of CO2')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = pd.DataFrame({
    'carbon_intensity': [200, 300, 400, 350, 500],  # gCO2/kWh
    'energy_consumption': [400, 500, 450, 550, 600],  # kWh
})

# Calculate carbon footprint
data['carbon_footprint'] = data['carbon_intensity'] * data['energy_consumption']

# Scatter plot between Carbon Intensity and Energy Consumption
plt.figure(figsize=(8, 6))
plt.scatter(data['energy_consumption'], data['carbon_intensity'], color='blue')
plt.title('Energy Consumption vs Carbon Intensity')
plt.xlabel('Energy Consumption (kWh)')
plt.ylabel('Carbon Intensity (gCO2/kWh)')
plt.grid(True)
plt.show()

# Line plot of Carbon Footprint over Energy Consumption
plt.figure(figsize=(8, 6))
plt.plot(data['energy_consumption'], data['carbon_footprint'], marker='o', color='red')
plt.title('Energy Consumption vs Carbon Footprint')
plt.xlabel('Energy Consumption (kWh)')
plt.ylabel('Carbon Footprint (grams of CO2)')
plt.grid(True)
plt.show()

# Bar plot for Carbon Footprint comparison
plt.figure(figsize=(8, 6))
plt.bar(data['energy_consumption'], data['carbon_footprint'], color='green')
plt.title('Carbon Footprint for Different Energy Consumption Levels')
plt.xlabel('Energy Consumption (kWh)')
plt.ylabel('Carbon Footprint (grams of CO2)')
plt.show()
```
